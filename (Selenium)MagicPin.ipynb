```python
import time
import pandas as pd
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.by import By
```

```python
# Setting Browser as Chrome
driver = webdriver.Chrome()
# Loading url
driver.get('https://magicpin.in/New-Delhi/Paharganj/Restaurant/Eatfit/store/61a193/delivery/')
# Maximizing window size for better visibility
driver.maximize_window()
```

```python
# Collecting all hidden menus and revealing them by clicking all of them
drop_downs = driver.find_elements(By.XPATH, '//img[@alt="down arrow"]')
for i in drop_downs:
    i.click()
```

```python
# Grabbing all add buttons since there are no hidden menus
items = driver.find_elements(By.XPATH, '//button[@class="countCta add "]')
len(items)
# Decaled a list to storing all prices
prices = []
```

```python
# Looping through all 'add' buttons
for item in items:
    # Making sure that the button is not intercepted with any other thing on screen
    driver.execute_script("arguments[0].scrollIntoView({behavior: 'auto', block: 'center'});", item)
    item.click()
    try:
        # If a separate 'ADD' dialog box pops up
        cart = driver.find_element(By.XPATH, '//button[@class="addCTA"]')    
        cart.click()
        # Wait for proper load
        time.sleep(2)
    except NoSuchElementException as e:
        print("No Such Element Found")
    except Exception as e:    
        print("Error:", e)
    # Grabbing original price(initial_price) and discounted price(final_price)
    initial_price = driver.find_element(By.XPATH, '//span[@class="highPrice"]')
    final_price = driver.find_element(By.XPATH, '//span[@class="finalPrice"]')
    prices.append([initial_price.text,final_price.text])
    # Locating the '-' button to empty the present cart
    empty_cart = driver.find_element(By.XPATH, '//img[@class="countActions remove err-handled"]')
    empty_cart.click()
    time.sleep(2)
```

```python
# Storing Prices
df = pd.DataFrame(prices, columns=['OriginalPrice','DiscountPrice'])
df
```

```python
df.to_csv('Task2.csv')
```
